```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

### creating a list of words

```python
text=("Python Python Python Matplotlib Matplotlib Seaborn Network Plot Violin Chart Pandas Datascience Wordcloud Spider Radar Parrallel Alpha Color Brewer Density Scatter Barplot Barplot Boxplot Violinplot Treemap Stacked Area Chart Chart Visualization Dataviz Donut Pie Time-Series Wordcloud Wordcloud Sankey Bubble")
```

### Create a wordcloud object

```python
wordcloud= WordCloud(width=480, height=500, margin=0).generate(text)
```

### Adjusting the Font Size, changing background color and font color

```python
wordcloud= WordCloud(width=480, height=480, max_font_size=30, min_font_size=10,background_color="Purple", colormap="Blues").generate(text)
```

### display generated image

```python
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.margins(x=0, y=0)
plt.show()
```
