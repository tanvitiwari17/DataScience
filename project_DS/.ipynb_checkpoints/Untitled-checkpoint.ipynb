##### get_dummies()

```python
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns
```

```python
# since a person can be absent only because of one reason, checking for sum:
reason_columns['check']=reason_columns.sum(axis=1)
reason_columns
```
