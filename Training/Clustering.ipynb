## Clustering
An unsupervised algorithm:
* Target/Independent variable is not defined
* patterns in the observation based on their featires are found.

#### Properties of Clustering:
* All the data points in a cluster should be similar to each other.
* The data points from different clusters should be as different as possible.

#### Applications of Clustering:
* customer segmentation
* document clustering
* image segmentation
* recommendation engines


###  Evaluating Clustering Models

##### Inertia :
tells us how far the points within a cluster are. 

The distance of a point from the centroid withing the cluster is called Intra Cluster Distance,

Inertia gives the sum of Intra Cluster Distance

Lesser the inertia, better the test result.


##### Dunn Index:
Calculates the Intra Cluster Distance as well as takes in account Inter Cluster Distance


###### Dunn Index = min(inter cluster distance)/max(intra cluster distance)

The higher the value of dunn index, the better the clusters are.

## K-means

k-means is a centroid based or distance-based algorithm

each cluster is defined by its centroid

* The ojective is to minimize the sum of distances between the points and their respective cluster centroid

##### Stopping Criteria
* if the centroids of newly formed clusters are not changing.
* If the points are not changing their clusters even after training the algorithm for multiple iterations.
* if the maximum number of iterations are reached.

##### Measuring Distance
 
#### * Euclidean Distance
used to measure distance between the points and the clusters.


##### Normalization: We subtract the means of the variables and divide by their standard deviation.

Scale data before appling k means to give equal weightage to all data points.

### Challenges in Implementing K-Means

* varied sizes of clusters
* varied densities of original points

#### solution:
* use more clusters
* merge the closest clusters until a reuired number of clusters are obtained.


maximum possible clusters = number of observations 
* with increase in number of clusters, computation cost increases
if high computational resources are not available, choose a lesser number of clusters.
