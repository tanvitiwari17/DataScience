## Ensemble Models
-> combining diverse models to generate better results

Instead of using a single model, use group of models and based on the predictions of each model, make a final decision

##### Pros:
* capture most of the diverse signals or patterns
* less incorrect predictions
* reduce overfitting
* performance

##### Cons:
* Increase complexity
* not interpretable as a whole
* increase time complexity and computational requirement.


### Bootstrap Sampling
Each sample is given a random sample from original data with replacement

### Bagging 
It focuses on bringing the diversity in the model using diverse data

### Random Forest:
* special case of bagging.
* Decision tree is used as the base model.
* One of the most popular ensemble algorithms.
* performs a random sampling of data points and features.

##### The number of bootstrap samples = number of models to trains


The data point/row sampling is done at the tree level

feature sampling is donw at the split level.
