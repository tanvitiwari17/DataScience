### Decision Tree
handles both _Regression_ and _Classification_ Problems

Unlike Linear and Logistic Regression (that makes restrictions or assumptions), Decision Tree does no make restrictions and assumptions.


Predictive Models: PAramentric and Non-Parametric

#### Parametric Models:
* Make strong assumtions about the form of mapping function
* very simple and interpretable
* Set of parameters does not depend upon the amount of data



###### Steps:
1. Select form of function
2. Learn the coefficients/parameters for the function from the training data.. 

_Linear Regression and Logistic Regression falls into the category of Parametric Algorithm_





#### Non Parametric Models:
* Do not make strong assumtions about the form of mapping function
* Free to learn any functional form from the training data
* Best fit the training data in constructing the mapping function
* Fit a large number of functional forms


_Desicion Tree fall into the category of Non Parametric Algorithm_




# Types of Decision Trees:

1. Classification Decision Tree (target : categorical)
2. Regression Decision Tree (target : continuous)


## Decision tree Classifier
### How to split nodes:
* Gini Impurity
* Information gain using entropy

## Gini Impurity: 
Measures the impurity of the nodes.
##### Gini impurity = 1-gini
gini ranges from 0 to 1
higher the value, more pure the nodes


Lower the gini impurity, more pure nodes

* gini only works with categorical targets

##### Steps to calculate gini impurity
1. Calculate the gini impurity for sub nodes


2. Calculate te gini impurity for split.


##### The split producing minimum gini impurity is selected as the first split

## Information Gain
###### The difference between information needed to describe the parent node and information needed to describe the children node
* works only with categorical data
The more impure the node is, the more information is required to describe them

The more is the homogeneity of the nodes, the more is the information gain and vice-versa.

###### IG = 1-Entropy(subnodes)


* lower entropy -> pure nodes -> more information gain
* high entropy -> less pure nodes -> less information gain



#### lesser entropy -> higher IG -> more homogeneity -> more purity of nodes

## Decision Tree Regressor

information gaina and gini impurity are not applicable to decision tree regressor to calculate impurity
This is because the target value in a decision tree regressor is a continuous value.

The variable with lower variance is used for splitting.
